```python
def extract_value(s):
    if s.count('/') == 2:
        return s.split('/')[1]
    else:
        return s 
    
concatenated_df['number_of_open_complaints'] = concatenated_df['number_of_open_complaints'].astype(str)
    
concatenated_df['number_of_open_complaints'] = concatenated_df['number_of_open_complaints'].apply(extract_value)
concatenated_df
```

```python
concatenated_df['customer_lifetime_value'] = concatenated_df['customer_lifetime_value'].astype(float)
concatenated_df['customer_lifetime_value'] = concatenated_df['customer_lifetime_value'].astype(float)

result = concatenated_df.dtypes
print(result)
```

```python
concatenated_df['gender'].unique()
```

```python
concatenated_df['gender'] = concatenated_df['gender'].str.replace('Femal', 'F')
concatenated_df['gender'] = concatenated_df['gender'].str.replace('female', 'F')
concatenated_df['gender'] = concatenated_df['gender'].str.replace('Male', 'M')
print(concatenated_df)
```

```python
concatenated_df['state'].unique()
```

```python
concatenated_df['state'] = concatenated_df['state'].str.replace('AZ', 'Arizona')
concatenated_df['state'] = concatenated_df['state'].str.replace('WA', 'Washington')
concatenated_df['state'] = concatenated_df['state'].str.replace('Cali', 'California')
concatenated_df['state'] = concatenated_df['state'].str.replace('Californiafornia', 'California')
print(concatenated_df)
```

```python
#Replacing null values – Replace missing values with means of the column (for numerical columns). Pay attention that the Income feature for instance has 0s which is equivalent to null values. (We assume here that there is no such income with 0 as it refers to missing values) Hint: numpy.nan is considered of float64 data type.
concatenated_df['number_of_open_complaints'] = concatenated_df['number_of_open_complaints'].fillna(value=concatenated_df['number_of_open_complaints'].mean)
concatenated_df['number_of_open_complaints']
```

```python
concatenated_df['customer_lifetime_value'] = concatenated_df['customer_lifetime_value'].fillna(value=concatenated_df['customer_lifetime_value'].mean)
concatenated_df['customer_lifetime_value']
```

```python
concatenated_df['income'] = concatenated_df['income'].fillna(value=concatenated_df['income'].mean)
concatenated_df['income']
```

```python
concatenated_df['monthly_premium_auto'] = concatenated_df['monthly_premium_auto'].fillna(value=concatenated_df['monthly_premium_auto'].mean)
concatenated_df['monthly_premium_auto']
```

```python
def replace_state_with_zone(concatenated_df): 
    
    state_to_zone = {
        'California': 'West Region',
        'Oregon': 'North West',
        'Washington': 'East',
        'Arizona': 'Central',
        'Nevada': 'Central'
    }

    concatenated_df['state'] = concatenated_df['state'].map(state_to_zone)

    return concatenated_df

print(concatenated_df)
```
